# Random List Solution


```python
import random
lst = []
for i in range(10):
    lst.append(random.randint(-10,10))
lst
```

# Positve and Negative Number Print List Solution

```python
pos_list = []
neg_list = []
for i in lst:
    if i >= 0:
        pos_list.append(i)
    else:
        neg_list.append(i)
print('Positive Numbers: ', end='')
for i in range(len(pos_list)):
    if i != max(range(len(pos_list))):
        print(f'{pos_list[i]}, ', end='')
    else:
        print(f'{pos_list[i]}')
print('Negative Numbers: ', end='')
for i in range(len(neg_list)):
    if i != max(range(len(neg_list))):
        print(f'{neg_list[i]}, ', end='')
    else:
        print(f'{neg_list[i]}')
```

# All Multiples Solution with Challenge

```python
# Given a random positive number calculate what numbers it can be divisible by and print them with commas seperating them. For this exercise lets say it can't be greater than 100 or less than 1
random_num = random.randint(1,100)
print(f'The number {random_num} is divisible by: ', end='')
mult_list = []
for i in range(1,random_num+1):
    if random_num % i == 0:
        mult_list.append(i)
for i in mult_list:
    if i != max(mult_list):
        print(f'{i}, ', end='')
    else:
        print(f'{i}')
```

# Simple Calculator Solution



```python
yn = 'Y'
print('This calculator can only do +, -, /, and *')
while ((yn == 'Y') | (yn == 'y')):
    num1, sign, num2 = input('Give me an equation with spaces between numbers and signs(Example: 2 - 4): ').split()
    num1 = int(num1)
    num2 = int(num2)
    if (sign == '+'):
        print(f'The sum of the 2 numbers is: {num1 + num2}')
        yn = input('If you want to calculate something else, enter Y: ')
    elif (sign == '/'):
        print(f'The num1 divided by num2 is: {num1 / num2}')
        yn = input('If you want to calculate something else, enter Y: ')
    elif (sign == '*'):
        print(f'The num1 times num2 is: {num1 * num2}')
        yn = input('If you want to calculate something else, enter Y: ')
    elif (sign == '-'):
        print(f'The num1 minus num2 is: {num1 - num2}')
        yn = input('If you want to calculate something else, enter Y: ')
    else:
        print('Sign is not +, -, /, or *')
        yn = input('If you want to calculate something else, enter Y: ')
```
